```python
import pandas as pd
df = pd.read_csv('Employee Sample Data - A.csv')
```

Standardize Column Names from while spaces to --> '_'bold text

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df
```

Trim Whitespace

```python
# Loop through each column
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].str.strip()
df
```

 Case Normalization

```python
df['full_name'] = df['full_name'].str.title()
df['job_title'] = df['job_title'].str.title()
df['city'] = df['city'].str.title()

df['department'] = df['department'].str.upper()
df['business_unit'] = df['business_unit'].str.upper()
df['country'] = df['country'].str.upper()
df['gender'] = df['gender'].str.upper()
```

Data Type Conversions

```python
df['hire_date'] = pd.to_datetime(df['hire_date'], errors='coerce')
df['exit_date'] = pd.to_datetime(df['exit_date'], errors='coerce')
print(df.dtypes)

df
```

Handling Missing Values

```python
# Drop rows where 'full_name', 'hire_date', or 'job_title' are missing
df = df.dropna(subset=['full_name', 'hire_date', 'job_title'])

# Clean 'annual_salary' to ensure it's numeric
df['annual_salary'] = pd.to_numeric(df['annual_salary'], errors='coerce')

# Clean 'bonus_%': remove % sign if needed and convert to numeric
df['bonus_%'] = df['bonus_%'].astype(str).str.replace('%', '', regex=False)
df['bonus_%'] = pd.to_numeric(df['bonus_%'], errors='coerce')

# Fill missing categorical values with 'UNKNOWN'
df['department'] = df['department'].fillna('UNKNOWN')
df['gender'] = df['gender'].fillna('UNKNOWN')
df['ethnicity'] = df['ethnicity'].fillna('UNKNOWN')
df['country'] = df['country'].fillna('UNKNOWN')
df['city'] = df['city'].fillna('UNKNOWN')

# Fill missing numeric values with the median of each column
df['age'] = df['age'].fillna(df['age'].median())
df['annual_salary'] = df['annual_salary'].fillna(df['annual_salary'].median())
df['bonus_%'] = df['bonus_%'].fillna(df['bonus_%'].median())

# Final check to confirm only 'exit_date' has missing values now
print(df.isna().sum())

```

```python
# Grouping by 'department' and calculating the mean of 'age' and 'annual_salary'
result = df.groupby('department').agg({
    'age': 'mean',
    'annual_salary': 'mean'
})

# Rename columns for better readability in the output
result = result.rename(columns={'age': 'Average Age', 'annual_salary': 'Average Salary'})

print(result)

# Alternatively, using the same lowercase column names with the simpler syntax
result = df.groupby('department')[['age', 'annual_salary']].mean()
print(result)
```

```python
# Grouping by 'department' and 'ethnicity' to get min, max age and median salary
result = df.groupby(['department', 'ethnicity']).agg({
    'age': ['min', 'max'],
    'annual_salary': 'median'
})

# Rename the resulting columns for readability
result.columns = ['Min Age', 'Max Age', 'Median Salary']

# Display the cleaned-up result
print(result)
```
